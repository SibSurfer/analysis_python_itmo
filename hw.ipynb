```python
from typing import Optional

class Chocolate:
  def __init__(self, price: float, topping: Optional[str] = None):
    if price <= 0:
      raise ValueError("price should be a positive number")
    self.price = price
    self.topping = topping

  def show_my_chocolate(self) -> None:
    if self.topping is not None:
      print(f"Шоколадный батончик, добавка: {self.topping}")
    else:
      print("Обычный шоколадный батончик")

  def __str__(self) -> str:
    return f"Шоколадный батончик, {self.price} рублей"
```

```python
expensive_choco_without_top = Chocolate(230)
expensive_choco_without_top.show_my_chocolate()
print(expensive_choco_without_top)
```

```python
cheap_choco_with_strawberry = Chocolate(89, "клубника")
cheap_choco_with_strawberry.show_my_chocolate()
print(cheap_choco_with_strawberry)
```
